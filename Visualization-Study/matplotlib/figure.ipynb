```
#참조 https://wikidocs.net/4761

import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도
```

```
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)  # 2행 2열의 첫번째
ax2 = fig.add_subplot(2, 2, 2)  # 2행 2열의 두번째
ax3 = fig.add_subplot(2, 2, 3)  # 2행 2열의 세번째
plt.show()
```

```
#subplot은 그래프의 axis 같은 개념

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

x = range(0, 100)   # 0~99 까지
y = [v*v for v in x]# x의 배열 제곱의 리스트

ax1.plot(x, y) 
ax2.bar(x, y)

plt.show()
```

```
x = np.arange(0.0, 2 * np.pi, 0.1)

fig = plt.figure()
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

ax1.plot(x, np.sin(x), 'b--') # sin x
ax2.plot(x, np.cos(x), 'r--') # cos x

plt.show()
```
