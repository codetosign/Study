```
#참조 https://matplotlib.org/3.3.3/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py

import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도
```

```
print(plt.plot([1, 2, 3, 4], [1, 4, 9, 16])) #선형
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 'ro'= 점찍기
plt.axis([0, 6, 0, 10]) # x축 시작값, x축 끝값, y축 시작값, y축 끝값

plt.xlabel('A fundamental consideration') # 근본에 대한 고찰
plt.ylabel('Quality of life') # 삶의 질
plt.show()
```
