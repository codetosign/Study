```
#참조 https://matplotlib.org/3.3.3/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py

import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib as mpl  # 기본 설정 만지는 용도
import matplotlib.pyplot as plt  # 그래프 그리는 용도
import matplotlib.font_manager as fm  # 폰트 관련 용도
```

```
print(plt.plot([1, 2, 3, 4], [1, 4, 9, 16])) #선형
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') # 'ro'= 빨간(r) 점찍기(o)
plt.axis([0, 6, 0, 10]) # x축 시작값, x축 끝값, y축 시작값, y축 끝값

plt.xlabel('A fundamental consideration') # 근본적 고찰
plt.ylabel('Quality of life') # 삶의 질
plt.show()
```

```
# np.arange 참조 -> https://codepractice.tistory.com/88
t = np.arange(0., 5., 0.2) # 5 전까지 0.2 씩 올리기

plt.plot(t, t, 'r--',   # red dashes
         t, t**2, 'bs', # blue squares      2 제곱
         t, t**3, 'g^') # green triangles   3 제곱
plt.show()
```

```
# np.randint 참조 -> https://nittaku.tistory.com/443

data = {'a': np.arange(50), # 0 ~ 49
        'c': np.random.randint(0, 50, 50),  # 0 ~ 49 사이의 임의의 랜덤값 50개
        'd': np.random.randn(50)} #가우시안 분포 값 50개
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data) # x의 위치, y의 위치, 색깔, 면적 what is data?
plt.show()
```

```
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100] 

plt.figure(figsize=(19, 3)) # 그래프 크기 가로, 세로

plt.subplot(131)  # 1행 3열 1번째
plt.bar(names, values)    # 막대

plt.subplot(132)  # 1행 3열 2번째
plt.scatter(names, values)# 점묘

plt.subplot(133)  # 1행 3열 3번째
plt.plot(names, values)   # 선형
plt.suptitle('Categorical Plotting')  # 3개의 그래프 윗상단 이름
plt.show()
```

```
def f(t):
    return (np.exp(-t)    # 자연상수 e의 마이너스 지수 (파라미터 값)승
     * np.cos(2*np.pi*t)) # 코사인 (2*파이 * 파라미터 값)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), 'bo', t2, f(t2), 'k') # 파란점형, 선형

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), 'r--') # 빨간색 대쉬
plt.show()
```

```
import matplotlib.pyplot as plt
plt.figure(1)                # the first figure
plt.subplot(211)             # the first subplot in the first figure
plt.plot([1, 2, 3])
plt.subplot(212)             # the second subplot in the first figure
plt.plot([4, 5, 6])


plt.figure(2)                # a second figure
plt.plot([4, 5, 6])          # creates a subplot(111) by default

plt.figure(1)                # figure 1 current; subplot(212) still current
plt.subplot(211)             # make subplot(211) in figure1 current
plt.title('Easy as 1, 2, 3') # subplot 211 title
```

```
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# the histogram of the data
n, bins, patches = plt.hist(x, 50, density=1, facecolor='violet', alpha=1)
# density > 밀도 / alpha > 투명도

plt.xlabel('Smarts', fontsize=15, color= 'red')  # xlabel 옵션
plt.ylabel('Probability', fontsize= 15, color= 'blue')
plt.title('Histogram of IQ', fontsize= 30)  # 상단 타이틀
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')  # 글자 삽입
plt.axis([40, 160, 0, 0.03])  # xmin, xmax, ymin, ymax
plt.grid(True)  # grid 생성
plt.show()
```
